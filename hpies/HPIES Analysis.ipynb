```python
def get_scalar(values):
    values = set(values)
    assert len(values) == 1
    return values.pop()

def compute_hef_demod(half_cycles, window_index=0, window_len=3, vref=2.5, preamp=800, divider=0.1773, tskip=15):
    npp = []
    hcno = []
    uxt = []
    ind = []
    abu = []
    e1a = []
    e1b = []
    e1c = []
    e2a = []
    e2b = []
    e2c = []
    sqw = []
    wts = []
    
    for index, half_cycle in half_cycles[window_index:window_index+window_len]:
        n = len(half_cycle.time)
        times = half_cycle.time.values.astype('f64') / 1e9
        this_hcno = get_scalar(half_cycle.hpies_hcno.values)
        this_abu = get_scalar(half_cycle.hpies_dest.values)
        
        # offset and scale
        off_ef = 2**15 * half_cycle.hpies_navg_ef
        sf_ef = vref / 2**16 / preamp / divider * 1e6 / half_cycle.hpies_navg_ef
        
        npp = np.append(npp, n)
        uxt = np.append(uxt, times)
        hcno = np.append(hcno, this_hcno)

        abu = np.append(abu, this_abu)
        e1a = np.append(e1a, (half_cycle.hpies_e1a.values - off_ef) * sf_ef)
        e1b = np.append(e1b, -(half_cycle.hpies_e1b.values - off_ef) * sf_ef)
        e1c = np.append(e1c, (half_cycle.hpies_e1c.values - off_ef) * sf_ef)
        e2a = np.append(e2a, (half_cycle.hpies_e2a.values - off_ef) * sf_ef)
        e2b = np.append(e2b, -(half_cycle.hpies_e2b.values - off_ef) * sf_ef)
        e2c = np.append(e2c, (half_cycle.hpies_e2c.values - off_ef) * sf_ef)

        if this_abu == 'a':
            sqw = np.append(sqw, +0.5 * np.ones(n))
        if this_abu == 'b':
            sqw = np.append(sqw, -0.5 * np.ones(n))

        t = half_cycle.time.values - half_cycle.time.values[0]
        k = np.nonzero(t < tskip)[0]
        w = np.ones(n, dtype=np.int)
        w[k] = 0
        wts = np.append(wts, w)

    if not all(np.diff(hcno) == 1):
        return

    trnd = uxt - np.mean(uxt)
    trnd = trnd / (trnd[-1] - trnd[0])
    const = np.ones(len(trnd))

    ma = []
    mb = []
    for i in range(window_len):
        if i % 2 == 0:
            ma.append('b')
            mb.append('a')
        else:
            ma.append('a')
            mb.append('b')

    ju = np.nonzero(wts > 0)[0]

    if len(ju) > 3:
        uxtmean = np.mean(uxt[ju])
        uxtmin = uxt[ju[0]]
        uxtmax = uxt[ju[-1]]

        # basis matrix for least squares fitting using np.linalg.lstsq()
        Bju = np.vstack([sqw[ju], trnd[ju], const[ju]]).T

        if list(abu) == list(ma):
            abm = 'a'
            e1sqw, e1trnd, e1const = np.linalg.lstsq(Bju, e1a[ju])[0]
            e2sqw, e2trnd, e2const = np.linalg.lstsq(Bju, e2a[ju])[0]
            e1res = e1a[ju] - (e1sqw * sqw[ju] + e1trnd * trnd[ju] + e1const)
            e2res = e2a[ju] - (e2sqw * sqw[ju] + e2trnd * trnd[ju] + e2const)
        elif list(abu) == list(mb):
            abm = 'b'
            # flipped signs here because e1b and e2b signs flipped much earlier
            e1sqw, e1trnd, e1const = np.linalg.lstsq(Bju, -e1b[ju])[0]
            e2sqw, e2trnd, e2const = np.linalg.lstsq(Bju, -e2b[ju])[0]
            e1res = -e1b[ju] - (e1sqw * sqw[ju] + e1trnd * trnd[ju] + e1const)
            e2res = -e2b[ju] - (e2sqw * sqw[ju] + e2trnd * trnd[ju] + e2const)
        else:
            return None

        mf = int(window_len / 2)

        e1std = np.std(e1res)
        e2std = np.std(e2res)

        e1c_sqw, e1c_trnd, e1c_const = np.linalg.lstsq(Bju, e1c[ju])[0]
        e2c_sqw, e2c_trnd, e2c_const = np.linalg.lstsq(Bju, e2c[ju])[0]
        e1c_res = e1c[ju] - (e1c_sqw * sqw[ju] + e1c_trnd * trnd[ju] + e1c_const)
        e2c_res = e2c[ju] - (e2c_sqw * sqw[ju] + e2c_trnd * trnd[ju] + e2c_const)
        e1c_std = np.std(e1c_res)
        e2c_std = np.std(e2c_res)
        
        return (uxtmean, uxtmin, uxtmax, abm, e1sqw, e2sqw,
                e1const, e2const, e1std, e2std, e1c_sqw, e2c_sqw,
                e1c_std, e2c_std)
```

```python
results = []
for i in xrange(0, len(half_cycles)):
    r = compute_hef_demod(half_cycles, window_index=i)
    if r is not None:
        results.append(r)
    
df = pd.DataFrame(results, columns=('uxtmean', 'uxtmin', 'uxtmax', 'abm', 'e1sqw', 'e2sqw',
                                    'e1const', 'e2const', 'e1std', 'e2std', 'e1c_sqw', 'e2c_sqw',
                                    'e1c_std', 'e2c_std'))
df = df.set_index(pd.to_datetime(df.uxtmean, unit='s'))
```
