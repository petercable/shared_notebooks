```python
%matplotlib inline
# import seaborn as sbs
import xarray as xr
import pandas as pd
import datetime
import matplotlib

matplotlib.rcParams['figure.figsize'] = (15.0, 5.0)
matplotlib.rcParams['axes.formatter.useoffset'] = False

NTP_EPOCH = (datetime.datetime(1970,1,1) - datetime.datetime(1900,1,1)).total_seconds()

fp1 = '20161202T155718-RS03CCAL-MJ03F-05-BOTPTA301-playback-botpt_nano_15s_precomputed/' + \
  'deployment0000_RS03CCAL-MJ03F-05-BOTPTA301-playback-botpt_nano_15s_precomputed_20140829T205945-20150519T221830.nc'
    
fp2 = '20161207T132408-RS03ECAL-MJ03E-06-BOTPTA302-playback-botpt_nano_15s_precomputed/' + \
  'deployment0000_RS03ECAL-MJ03E-06-BOTPTA302-playback-botpt_nano_15s_precomputed_20140903T220515-20150519T221745.nc'
    
fp3 = '20161207T132236-RS03INT2-MJ03D-06-BOTPTA303-playback-botpt_nano_15s_precomputed/' + \
  'deployment0000_RS03INT2-MJ03D-06-BOTPTA303-playback-botpt_nano_15s_precomputed_20140829T231730-20150519T221645.nc'
    
def get_dataframe(fname):
    dataframe = xr.open_dataset(fname, decode_times=False).to_dataframe()
    dataframe['time'] = pd.to_datetime(dataframe.botsflu_time15s.values - NTP_EPOCH, unit='s')
    dataframe = dataframe.set_index('time')
    return dataframe
```

```python
df1 = get_dataframe(fp1)
df2 = get_dataframe(fp2)
df3 = get_dataframe(fp3)
```
